```python
filtered_dict_data = {"description": [], "estate_type": [], "square": [], "price": []}
num_samples = 0
for i, data in enumerate(json_data):
    num_samples += 1
    for k, v in filtered_dict_data.items():
        v.append(data[k])
```

```python
def split_data(dict_data, size, train_ratio=0.82):
    test_indices = np.zeros((size,), dtype=bool)
    for i, v in enumerate(dict_data["price"]):
        if isinstance(v, float):
            if v > 1e13 or v < 10:
                test_indices[i] = True
        else:
            test_indices[i] = True

    random_split = np.random.uniform(size=(num_samples,)) < train_ratio

    train_indices = random_split & (~test_indices)
    validation_indices = (~random_split) & (~test_indices)

    train_dict = {"description": [], "estate_type": [], "square": [], "price": []}
    validation_dict = {"description": [], "estate_type": [], "square": [], "price": []}
    test_dict = {"description": [], "estate_type": [], "square": [], "price": []}

    for i in range(size):
        if train_indices[i]:
            for k, v in train_dict.items():
                v.append(dict_data[k][i])
        if validation_indices[i]:
            for k, v in validation_dict.items():
                v.append(dict_data[k][i])
        if test_indices[i]:
            for k, v in test_dict.items():
                v.append(dict_data[k][i])
    return {
        "train": train_dict,
        "validation": validation_dict,
        "test": test_dict
    }
```

```python
splited_data = split_data(filtered_dict_data, num_samples)
```

```python
train_dict = splited_data["train"]
validation_dict = splited_data["validation"]
test_dict = splited_data["test"]
```

```python
with open(data_dir + "train_dict.json", "w", encoding="utf-8") as train_writer:
    json.dump(train_dict, train_writer)

with open(data_dir + "validation_dict.json", "w", encoding="utf-8") as val_writer:
    json.dump(validation_dict, val_writer)

with open(data_dir + "test_dict.json", "w", encoding="utf-8") as test_writer:
    json.dump(test_dict, test_writer)
```
